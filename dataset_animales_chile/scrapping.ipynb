```python
import requests
import pandas as pd
from tqdm import tqdm

BASE_URL = "https://api.inaturalist.org/v1/taxa"
PARAMS = {
    "place_id": 6744,      # Chile
    "rank": "species",
    "per_page": 200,
    "page": 1
}

data = []

print("🔍 Obteniendo especies desde iNaturalist...")
while True:
    r = requests.get(BASE_URL, params=PARAMS)
    if r.status_code != 200:
        break

    results = r.json()["results"]
    if not results:
        break

    for item in results:
        data.append({
            "taxon_id": item.get("id"),
            "nombre_cientifico": item.get("name"),
            "nombre_comun": item.get("preferred_common_name"),
            "iconic_taxon_name": item.get("iconic_taxon_name"),
            "ruta": item["default_photo"]["url"] if item.get("default_photo") else None
        })

    PARAMS["page"] += 1

df = pd.DataFrame(data)
df.to_csv("dataset_animales_chile_completo.csv", index=False)
print(f"✅ CSV generado con {len(df)} especies únicas")
print(df.head())
```

```python
import os
import requests
import pandas as pd
from tqdm import tqdm

CSV_PATH = "dataset_animales_chile_completo.csv"
DESTINO = "dataset_animales_chile"
VALID_CLASSES = ['Actinopterygii', 'Amphibia', 'Arachnida', 'Aves', 'Insecta', 'Mammalia', 'Reptilia']

# Crear carpeta base si no existe
os.makedirs(DESTINO, exist_ok=True)

df = pd.read_csv(CSV_PATH)

for clase in VALID_CLASSES:
    clase_dir = os.path.join(DESTINO, clase)
    os.makedirs(clase_dir, exist_ok=True)

    urls = df[df["iconic_taxon_name"] == clase]["ruta"].dropna().unique()
    print(f"📥 Descargando {len(urls)} imágenes para {clase}...")

    for i, url in enumerate(tqdm(urls)):
        try:
            ext = url.split('.')[-1].split('?')[0].lower()
            if ext not in ["jpg", "jpeg", "png"]:
                ext = "jpg"

            img_path = os.path.join(clase_dir, f"{clase}_{i}.{ext}")

            if not os.path.exists(img_path):  # No volver a descargar si existe
                img_data = requests.get(url, timeout=10).content
                with open(img_path, 'wb') as f:
                    f.write(img_data)
        except Exception:
            pass

print("\n✅ Dataset completo descargado en:", DESTINO)
```

```python
import os
import shutil
import random

ORIGEN = "dataset_animales_chile"
DESTINO = "dataset_animales_mini"
IMAGES_PER_CLASS = 50

# Eliminar mini-dataset previo si existe
if os.path.exists(DESTINO):
    shutil.rmtree(DESTINO)
os.makedirs(DESTINO)

for clase in os.listdir(ORIGEN):
    origen_clase = os.path.join(ORIGEN, clase)
    if os.path.isdir(origen_clase):
        destino_clase = os.path.join(DESTINO, clase)
        os.makedirs(destino_clase, exist_ok=True)
        
        imagenes = [f for f in os.listdir(origen_clase) if f.lower().endswith((".jpg", ".jpeg", ".png"))]
        random.shuffle(imagenes)
        seleccionadas = imagenes[:min(IMAGES_PER_CLASS, len(imagenes))]
        
        for img in seleccionadas:
            shutil.copy(os.path.join(origen_clase, img), os.path.join(destino_clase, img))
        
        print(f"✅ {clase}: {len(seleccionadas)} imágenes copiadas")

print("\n🎉 Mini-dataset listo en:", DESTINO)
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.applications.mobilenet_v2 import preprocess_input
from tensorflow.keras import layers, models
from tensorflow.keras.callbacks import EarlyStopping
import os

# ===== CONFIGURACIÓN =====
DATASET_DIR = "dataset_animales_mini"
IMG_SIZE = (224, 224)
BATCH_SIZE = 16
EPOCHS = 10

# ===== GENERADORES =====
train_datagen = ImageDataGenerator(
    preprocessing_function=preprocess_input,
    validation_split=0.2
)

train_generator = train_datagen.flow_from_directory(
    DATASET_DIR,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    subset="training",
    seed=42
)

val_generator = train_datagen.flow_from_directory(
    DATASET_DIR,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    subset="validation",
    seed=42
)

print(f"✅ Clases detectadas: {train_generator.class_indices}")
print(f"✅ Imágenes entrenamiento: {train_generator.samples}")
print(f"✅ Imágenes validación: {val_generator.samples}")

# ===== MODELO BASE =====
base_model = MobileNetV2(input_shape=(224, 224, 3), include_top=False, weights="imagenet")
base_model.trainable = False  # Fase 1: Congelado

modelo = models.Sequential([
    base_model,
    layers.GlobalAveragePooling2D(),
    layers.Dense(128, activation='relu'),
    layers.Dropout(0.3),
    layers.Dense(train_generator.num_classes, activation='softmax')
])

modelo.compile(optimizer='adam',
               loss='categorical_crossentropy',
               metrics=['accuracy'])

# ===== ENTRENAMIENTO INICIAL =====
early_stop = EarlyStopping(monitor="val_loss", patience=3, restore_best_weights=True)
history = modelo.fit(train_generator, validation_data=val_generator, epochs=EPOCHS, callbacks=[early_stop])

# ===== FINE-TUNING =====
base_model.trainable = True
modelo.compile(optimizer=tf.keras.optimizers.Adam(1e-5),
               loss='categorical_crossentropy',
               metrics=['accuracy'])

history_finetune = modelo.fit(train_generator, validation_data=val_generator, epochs=5, callbacks=[early_stop])

# ===== GUARDAR MODELO =====
modelo.save("modelo_nivel1_mobilenetv2_mini.h5")
print("✅ Modelo entrenado y guardado como modelo_nivel1_mobilenetv2_mini.h5")
```

```python
import numpy as np
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.mobilenet_v2 import preprocess_input
from tensorflow.keras.models import load_model

# ===== CARGAR MODELO =====
modelo = load_model("modelo_nivel1_mobilenetv2_mini.h5")

clases = {
    0: 'Actinopterygii',
    1: 'Amphibia',
    2: 'Arachnida',
    3: 'Aves',
    4: 'Insecta',
    5: 'Mammalia',
    6: 'Reptilia'
}

def predecir_imagen(ruta_imagen):
    img = image.load_img(ruta_imagen, target_size=(224, 224))
    x = image.img_to_array(img)
    x = np.expand_dims(x, axis=0)
    x = preprocess_input(x)

    pred = modelo.predict(x)[0]
    clase_predicha = np.argmax(pred)
    confianza = pred[clase_predicha] * 100

    print(f"✅ Predicción: {clases[clase_predicha]} ({confianza:.2f}% confianza)")
    return clases[clase_predicha], confianza
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping

import os

# ======== AUGMENTACIÓN FUERTE =========
train_datagen = ImageDataGenerator(
    rescale=1.0/255,
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.3,
    brightness_range=[0.7, 1.3],
    horizontal_flip=True,
    fill_mode='nearest',
    validation_split=0.2
)

val_datagen = ImageDataGenerator(rescale=1.0/255, validation_split=0.2)

# ======== DATASET MINI =========
DATASET_DIR = "dataset_animales_mini"

train_generator = train_datagen.flow_from_directory(
    DATASET_DIR,
    target_size=(224, 224),
    batch_size=16,
    class_mode='categorical',
    subset='training'
)

val_generator = val_datagen.flow_from_directory(
    DATASET_DIR,
    target_size=(224, 224),
    batch_size=16,
    class_mode='categorical',
    subset='validation'
)

print(f"✅ Clases detectadas: {train_generator.class_indices}")
print(f"✅ Imágenes entrenamiento: {train_generator.samples}")
print(f"✅ Imágenes validación: {val_generator.samples}")

# ======== MODELO BASE =========
base_model = MobileNetV2(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
base_model.trainable = False

x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dense(256, activation='relu')(x)
x = Dropout(0.4)(x)
preds = Dense(train_generator.num_classes, activation='softmax')(x)

modelo_mini = Model(inputs=base_model.input, outputs=preds)
modelo_mini.compile(optimizer=Adam(learning_rate=0.0005), loss='categorical_crossentropy', metrics=['accuracy'])

# ======== ENTRENAMIENTO INICIAL (CONGELADO) =========
early_stop = EarlyStopping(monitor="val_loss", patience=5, restore_best_weights=True)

history = modelo_mini.fit(
    train_generator,
    validation_data=val_generator,
    epochs=20,
    callbacks=[early_stop]
)

# ======== FINE-TUNING =========
base_model.trainable = True
for layer in base_model.layers[:-50]:
    layer.trainable = False

modelo_mini.compile(optimizer=Adam(learning_rate=0.00005), loss='categorical_crossentropy', metrics=['accuracy'])

history_finetune = modelo_mini.fit(
    train_generator,
    validation_data=val_generator,
    epochs=10,
    callbacks=[early_stop]
)

modelo_mini.save("modelo_nivel1_mobilenetv2_mini_fuerte.h5")
print("✅ Modelo entrenado y guardado como modelo_nivel1_mobilenetv2_mini_fuerte.h5")
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping

# ================================
# 🔥 DATA AUGMENTATION FUERTE
# ================================
train_datagen = ImageDataGenerator(
    rescale=1.0/255,
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.3,
    brightness_range=[0.7, 1.3],
    horizontal_flip=True,
    fill_mode='nearest',
    validation_split=0.2
)

val_datagen = ImageDataGenerator(rescale=1.0/255, validation_split=0.2)

DATASET_DIR = "dataset_animales_chile"

train_generator = train_datagen.flow_from_directory(
    DATASET_DIR,
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical',
    subset='training'
)

val_generator = val_datagen.flow_from_directory(
    DATASET_DIR,
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical',
    subset='validation'
)

print(f"✅ Clases detectadas: {train_generator.class_indices}")
print(f"✅ Imágenes entrenamiento: {train_generator.samples}")
print(f"✅ Imágenes validación: {val_generator.samples}")

# ================================
# 🔥 MODELO BASE
# ================================
base_model = MobileNetV2(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
base_model.trainable = False

x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dense(512, activation='relu')(x)
x = Dropout(0.4)(x)
preds = Dense(train_generator.num_classes, activation='softmax')(x)

modelo_full = Model(inputs=base_model.input, outputs=preds)
modelo_full.compile(optimizer=Adam(learning_rate=0.0005), loss='categorical_crossentropy', metrics=['accuracy'])

# ================================
# 🔥 ENTRENAMIENTO INICIAL (30 Épocas)
# ================================
early_stop = EarlyStopping(monitor="val_loss", patience=8, restore_best_weights=True)

history = modelo_full.fit(
    train_generator,
    validation_data=val_generator,
    epochs=30,
    callbacks=[early_stop]
)

# ================================
# 🔥 FINE-TUNING (20 Épocas)
# ================================
base_model.trainable = True
for layer in base_model.layers[:-50]:
    layer.trainable = False

modelo_full.compile(optimizer=Adam(learning_rate=0.00005), loss='categorical_crossentropy', metrics=['accuracy'])

history_finetune = modelo_full.fit(
    train_generator,
    validation_data=val_generator,
    epochs=20,
    callbacks=[early_stop]
)

modelo_full.save("modelo_nivel1_mobilenetv2_full.h5")
print("✅ Modelo completo entrenado y guardado como modelo_nivel1_mobilenetv2_full.h5")
```
